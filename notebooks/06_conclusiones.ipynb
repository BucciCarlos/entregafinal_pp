### 1. Resumen de la Fase Exploratoria (EDA)
* Línea Base Crítica: El albergue opera con una tasa histórica de cancelación del 37.24%, lo que representa una ineficiencia operativa severa y un alto costo de oportunidad.

* Carga Social: La demanda está fuertemente liderada por el segmento "Afiliados". Sin embargo, el análisis multivariado demostró que los grupos familiares grandes (3 a 4 adultos) absorben una proporción significativa del presupuesto de subsidios institucionales.

* Estacionalidad: Se detectaron picos claros en el costo de oportunidad perdido durante meses específicos, lo que exige abandonar las políticas de reserva estáticas durante todo el año.

### 2. Rendimiento del Modelado Predictivo
Se implementó un modelo de Machine Learning (Random Forest Optimizado mediante Cross-Validation) para predecir cancelaciones antes de que ocurran.

Mejora del Rendimiento: La optimización de hiperparámetros incrementó la precisión de detección de cancelaciones de un 50% a un 65%.

Segmentación de Riesgo: El modelo logró clasificar las reservas futuras en tres segmentos accionables:

* Riesgo Bajo (<35% prob.): Tasa real de cancelación mínima.

* Riesgo Medio (35%-65% prob.): Incertidumbre operativa moderada.

* Riesgo Alto (>65% prob.): El 76% de estas reservas terminan en cancelación efectiva.

### 3. Recomendaciones de Negocio y Plan de Acción
A partir de los datos empíricos, se recomienda a la dirección del albergue implementar las siguientes medidas:

* Política Dinámica de Overbooking: Autorizar una sobreventa moderada exclusivamente en las plazas ocupadas por perfiles clasificados como de Riesgo Alto, ya que la probabilidad empírica de no-show supera el 75%.

* Fricción Selectiva: Implementar cobros de seña por adelantado o reconfirmaciones telefónicas obligatorias a las 48hs solo para el segmento de Riesgo Medio/Alto. El segmento de Riesgo Bajo debe tener un flujo de check-in sin fricciones.

* Revisión de Canales: Renegociar cupos con los canales de origen que mostraron una tasa de cancelación desproporcionada respecto a los canales directos del gremio/UNSE.
